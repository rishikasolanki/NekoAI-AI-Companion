```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import numpy as np
import pandas as pd

def evaluate_model(name, y_true, y_pred):

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    print("="*40)
    print(name)
    print("="*40)
    print("MAE :", round(mae,3))
    print("MSE :", round(mse,3))
    print("RMSE:", round(rmse,3))
    print("R²  :", round(r2,3))
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

import numpy as np
```

```python
def evaluate_model(model_name, y_true, y_pred):

    mae = mean_absolute_error(y_true, y_pred)

    mse = mean_squared_error(y_true, y_pred)

    rmse = np.sqrt(mse)

    r2 = r2_score(y_true, y_pred)

    print("=" * 50)
    print("Model :", model_name)
    print("=" * 50)
    print("MAE :", round(mae,3))
    print("MSE :", round(mse,3))
    print("RMSE:", round(rmse,3))
    print("R² :", round(r2,3))
```

```python
evaluate_model(
    "Linear Regression",
    y_test,
    lr_pred
)

evaluate_model(
    "Decision Tree",
    y_test,
    dt_pred
)

evaluate_model(
    "Random Forest",
    y_test,
    rf_pred
)
```
